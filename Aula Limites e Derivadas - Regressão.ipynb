```python
# Import das bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dados simulados
# X variavel independente
# y variável dependente

X = np.array([[1,1], [1,2], [1,3], [1,4]])
y = np.array([2,4,5,4])
theta = np.zeros(2) # parametros do modelo, é um vetor de duas posições
alpha = 0.01 # Taxa de aprendizagem do modelo (pace de aprendizado)
iteractions = 1000 # quantidade de interações
```

```python
# Gradiente Descendente
m = len(y)
for _ in range(iteractions): # Intera sob a quantidade de interações que definimos
    h = np.dot(X, theta) # Calcula hipotese do modelo com base no theta atual
    error = h - y # Calcula o erro entre previsão (h) e valor real (y)
    gradient = np.dot(X.T, error) / m # Calcula o gradiente, direção de magnitude do erro. Diz como os parametros de theta devem ser ajustados
    theta -= alpha * gradient  # Atualiza os parametros de theta
```

```python
print(f'Parametros finais: {theta}')
```

```python
plt.scatter(X[:, 1], y, color="Blue")
plt.plot(X[:, 1], np.dot(X, theta), color="red")
plt.show()
```

```python
X = np.array([1,2,3,4,5])
y = np.array([2.2, 3.9, 5.1, 6.0, 7.2])
n = len(X)
b = 0 # Intercepto inicial
```

```python
# Testar valores de w (peso) em um intervalo mais amplo
w_values = np.linspace(-10, 10, 100)
j_values = [cost_function(w, X, y, b) for w in w_values]
```

```python
import numpy as np

# Dados fictícios
X = np.array([1, 2, 3, 4, 5])
y = np.array([2.2, 3.9, 5.1, 6.0, 7.2])
n = len(X)

# Função de gradiente descendente
def gradient_descent(X, y, alpha, max_iterations=10000, tolerance=0.01):
    w = 0.0
    b = 0.0
    iterations = 0
    converged = False

    for i in range(max_iterations):
        y_pred = w * X + b
        error = y_pred - y
        dw = (2/n) * np.sum(error * X)  # Derivada em relação a w
        db = (2/n) * np.sum(error)      # Derivada em relação a b

        w_new = w - alpha * dw
        b_new = b - alpha * db

        # Verificar convergência
        cost = (1/n) * np.sum((y - y_pred) ** 2)
        if cost < tolerance:
            converged = True
            break

        w, b = w_new, b_new
        iterations += 1

    return w, b, iterations, converged, cost

# Testar diferentes taxas de aprendizado
learning_rates = [0.1, 0.01, 0.001, 0.0001, 0.00015 ]
results = {}

for alpha in learning_rates:
    w, b, iterations, converged, final_cost = gradient_descent(X, y, alpha)
    results[alpha] = {'w': w, 'b': b, 'iterations': iterations, 'converged': converged, 'cost': final_cost}
    print(f"Taxa de aprendizado: {alpha}")
    print(f"  w: {w}, b: {b}")
    print(f"  Iterações: {iterations}, Convergiu: {converged}, Custo final: {final_cost}\n")
```

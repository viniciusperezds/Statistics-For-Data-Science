## Estatísticas

### Estatísticas dos clubes

- Geral

```
resumo_clubes = dados_carregados.groupby(['clube_id', 'clube']) \
                .agg(gols = pd.NamedAgg(column = 'G', aggfunc = 'sum'),
                     faltas = pd.NamedAgg(column = 'FC', aggfunc = 'sum'),
                     impedimentos = pd.NamedAgg(column = 'Impedimentos', aggfunc = 'sum')) \
                .sort_values('gols', ascending = False) \
                .reset_index()
```

```
resumo_clubes
```

```
resumo_clubes = dados_carregados.groupby(['clube_id', 'clube']) \
                                .agg(gols = pd.NamedAgg(column = 'G', aggfunc = 'sum'),
                                     faltas = pd.NamedAgg(column = 'FC', aggfunc = 'sum'),
                                     impedimentos = pd.NamedAgg(column = 'Impedimentos', aggfunc = 'sum')) \
                                .sort_values('gols', ascending = False) \
                                .reset_index()
```

```
resumo_clubes.head(3)
```

- Por rodada

```
resumo_clubes_rodada = dados_carregados.groupby(['rodada_id', 'clube_id', 'clube']) \
                                .agg(gols = pd.NamedAgg(column = 'G', aggfunc = 'sum'),
                                     faltas = pd.NamedAgg(column = 'FC', aggfunc = 'sum'),
                                     impedimentos = pd.NamedAgg(column = 'Impedimentos', aggfunc = 'sum')) \
                                .sort_values(['clube_id', 'rodada_id'], ascending = True) \
                                .reset_index()
```

```
resumo_clubes_rodada
```

### Estatísticas dos jogadores

```
resumo_jogadores = dados_carregados.groupby(['atleta_id', 'apelido']) \
                                   .agg(gols = pd.NamedAgg(column = 'G', aggfunc = 'sum'),
                                        faltas = pd.NamedAgg(column = 'FC', aggfunc = 'sum'),
                                        impedimentos = pd.NamedAgg(column = 'Impedimentos', aggfunc = 'sum')) \
                                   .sort_values('gols', ascending = False) \
                                   .reset_index()
```

```
resumo_jogadores.head()
```

```python
def formula_permutacao(n, r, repeticao = False):
  if repeticao is True:
    return n**r # Potencia  de elevado a r
  else:
    return factorial(n) / factorial(n-r)

def formula_combinacao(n, k, repeticao = False):
  if repeticao is False:
    return factorial(n) / (factorial(k) * factorial(n-k))
  else:
    return factorial(n+k-1) / (factorial(k) * factorial(n-1))
```

```python
formula_permutacao(3, 3, repeticao=True) # n=3 e r=3 com repetiçao = True
```

```python
formula_permutacao(3, 3, repeticao=False) # n=3 e r=3 com repetiçao = False
```

```python
formula_permutacao(n=10, # 10 maneiras de aranjar
                   r=6, # escolher 6 nymeros
                   repeticao = False # não repetir numero
                   )
```

```python
import matplotlib.pyplot as plt
import random

num_simulacoes = 10000
somas = []
ocorrencias_soma_7 = []

# Simulação dos lançamentos
for i in range(num_simulacoes):
    d1 = random.randint(1, 6)
    d2 = random.randint(1, 6)
    soma = d1 + d2
    somas.append(soma)

    if soma == 7:
        ocorrencias_soma_7.append(i)

# Calcular frequência experimental
freq_soma_7 = somas.count(7) / num_simulacoes
print("Probabilidade experimental da soma = 7:", freq_soma_7)

# Calcular os intervalos
intervalos = [ocorrencias_soma_7[i] - ocorrencias_soma_7[i-1] for i in range(1, len(ocorrencias_soma_7))]

# Calcular média dos intervalos
media_intervalo = sum(intervalos) / len(intervalos)
print("Média dos intervalos:", media_intervalo)

# Calcular média móvel com janela de 50
def media_movel(valores, janela=50):
    return [sum(valores[i:i+janela]) / janela for i in range(len(valores) - janela + 1)]

media_movel_50 = media_movel(intervalos, janela=50)

# Gráfico scatter + média + média móvel
plt.figure(figsize=(12, 6)) # tamanho da tela em branco
plt.scatter(range(1, len(intervalos) + 1), intervalos, color='purple', alpha=0.5, s=10, label='Intervalos')
plt.axhline(y=media_intervalo, color='red', linestyle='--', linewidth=2, label=f'Média ≈ {media_intervalo:.2f}')
plt.plot(range(25, len(media_movel_50) + 25), media_movel_50, color='orange', linewidth=2, label='Média Móvel (janela=50)')
plt.title("Intervalos entre Ocorrências da Soma 7 em Lançamentos de Dois Dados")
plt.xlabel("Ocorrência nº")
plt.ylabel("Intervalo (lançamentos entre somas 7)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
## Teorema de Bayes
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros
populacao = 100000
prob_doenca = 0.01
sensibilidade = 0.99  # P(+|D)
falso_positivo = 0.05  # P(+|¬D)

# Simulação da população
tem_doenca = np.random.rand(populacao) < prob_doenca  # True para doentes

# Simulação dos testes
testa_positivo = np.zeros(populacao, dtype=bool)

# Quem tem a doença: 99% chance de teste positivo
testa_positivo[tem_doenca] = np.random.rand(tem_doenca.sum()) < sensibilidade

# Quem NÃO tem a doença: 5% chance de falso positivo
testa_positivo[~tem_doenca] = np.random.rand((~tem_doenca).sum()) < falso_positivo

# Análise dos resultados
positivos = testa_positivo.sum()
positivos_e_doentes = np.logical_and(testa_positivo, tem_doenca).sum()

# Probabilidade empírica de estar doente dado que testou positivo
prob_bayes_empirica = positivos_e_doentes / positivos

print(f"Total de positivos: {positivos}")
print(f"Positivos e doentes: {positivos_e_doentes}")
print(f"Probabilidade P(doente | positivo): {prob_bayes_empirica:.4f} (~{prob_bayes_empirica*100:.2f}%)")

# Gráfico de barras
plt.bar(["Falso positivo", "Verdadeiro positivo"],
        [positivos - positivos_e_doentes, positivos_e_doentes],
        color=["orange", "green"])
plt.title("Resultado dos Testes Positivos")
plt.ylabel("Número de pessoas")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import random
from collections import Counter
import matplotlib.pyplot as plt

# Simular lançamentos
lancamentos = [random.randint(1, 6) for _ in range(100_000)]

# Contar frequência de cada número
contagem = Counter(lancamentos)
print("Contagem:", contagem)

# Plotar gráfico de barras
numeros = list(range(1, 7))
frequencias = [contagem[n] for n in numeros]

plt.figure(figsize=(8, 5))
plt.bar(numeros, frequencias, color='skyblue', edgecolor='black')
plt.title('Frequência dos Números em 100.000 Lançamentos de Dado')
plt.xlabel('Número do Dado')
plt.ylabel('Frequência')
plt.xticks(numeros)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# 5. Dada a função genérica de simulação de probabilidade abaixo:

def simular_probabilidade(evento, gerador, n=100_000):
    acertos = sum(1 for _ in range(n) if evento(gerador()))
    return acertos / n

# Aplique a função para:
#   a. Dado igual a 4
#   b. Soma de dois dados = 9
#   c. Pelo menos uma cara em 3 moedas
```

```python
def simular_probabilidade(evento, gerador, n=100_000):
    return sum(1 for _ in range(n) if evento(gerador())) / n

# a) Dado = 4
print("P(dado = 4):", simular_probabilidade(lambda x: x == 4, lambda: random.randint(1, 6)))

# b) Soma de 2 dados = 9
print("P(soma = 9):", simular_probabilidade(lambda x: sum(x) == 9, lambda: (random.randint(1,6), random.randint(1,6))))

# c) Pelo menos uma cara em 3 moedas
print("P(pelo menos uma cara):", simular_probabilidade(lambda x: any(x), lambda: [random.choice([0, 1]) for _ in range(3)]))
```

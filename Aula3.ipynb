## Classes

```python
class Cachorro:
    def __init__(self, nome, idade, raca):
        self.nome = nome
        self.idade = idade
        self.raca = raca

    def latir(self):
        return f"{self.nome} está latindo"

    def aniversario(self):
        self.idade += 1
        return f"{self.nome} agora tem {self.idade} anos"
```

```python
cachorro1 = Cachorro('Rex', 3, 'Labrador')
cachorro1.latir()
cachorro1.aniversario()
```

```python
cachorro1 = Cachorro('Bob', 9, 'Vira lata')
cachorro1.latir()
cachorro1.aniversario()
```

#### Simulando Dataset problematico

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
df = pd.DataFrame({
    'nome': ['Ana', 'João', 'ana', 'Maria', 'JOÃO'],
    'idade': [25, 30, np.nan, 22, np.nan],
    'salario': [3000, 4000, np.nan, 2800, 999999],
    'cidade': ['São Paulo', 'sao paulo', 'São Paulo', np.nan, 'RIO']
})

print("Original:")
print(df)
```

```python
df.info()
```

```python
# Padronizar os dados
df['nome'] = df['nome'].str.strip().str.lower()
df['cidade'] = df['cidade'].str.strip().str.lower()
df.head()
```

```python
# alterou os nomes incompletos
cidade_map = {
    'sao paulo': 'são paulo',
    'são paulo': 'são paulo',
    'rio': 'rio de janeiro'
}
df['cidade'] = df['cidade'].map(cidade_map)
```

```python
df.head()
```

```python
# remover dados duplicados
df.drop_duplicates(inplace=True)
```

```python
# Função isna() mostra os dados ausentes por coluna, se adicionar o .sum() traz a soma por coluna
df.isna().sum()
```

```python
#df['idade'].fillna(0, inplace=True)
#df.head()
```

```python
# Preencher dados faltantes
df['idade'].fillna(round(df['idade'].mean(),0), inplace=True)
df['salario'].fillna(df['salario'].mean(), inplace=True)
df['cidade'].fillna(df['cidade'].mode()[0], inplace=True)
```

```python
print(df['idade'].mean())
print(df['idade'].max())
print(df['idade'].min())
```

```python
df.head()
```

```python
## Normalização e padronização dos dados
scaler_norm = MinMaxScaler() # De 0 a 1
scaler_std = StandardScaler() # Z-score (padroniza dados com a media 0 e desvio 1)
```

```python
df['salario_normalizado'] = scaler_norm.fit_transform(df[['salario']])
df['salario_padronizado'] = scaler_std.fit_transform(df[['salario']])

df[['salario', 'salario_normalizado', 'salario_padronizado']].head()
```
